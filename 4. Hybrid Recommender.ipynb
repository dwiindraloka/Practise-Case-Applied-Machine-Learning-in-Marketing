## Import Library

```python
import pandas as pd
```

## Upload Dataset

### 1. Movies Detail

```python
data_movie = pd.read_csv("film1.csv")
data_movie.head()
```

### 2. People Rating

```python
data_rating = pd.read_csv("film.csv")
data_rating.drop("Timestamp", axis = 1, inplace = True)
data_rating = data_rating.fillna(0)
data_rating.head()
```

## Change Movies Detail's Column "Judul Film" into Index

```python
data_movie.set_index("Judul Film", inplace = True)
data_movie.head()
```

## Make Data Dummies for Movies Detail

```python
data_movie1 = pd.get_dummies(data_movie)
data_movie1.head()
```

## Normalize the Data Using MinMaxScaler for Movies Detail

```python
from sklearn.preprocessing import MinMaxScaler
data_movie2 = data_movie1.transpose()
col = data_movie1.columns
scaler = MinMaxScaler().fit_transform(data_movie1)
data_movie3 = pd.DataFrame(scaler, columns = col)
data_movie3.head()
```

## Calculate Each Movie's Similarity Using Cosine Similarity for Movies Detail

```python
from sklearn.metrics.pairwise import cosine_similarity
data_cosine = cosine_similarity(data_movie3)
data_cosine = pd.DataFrame(data_cosine, columns = data_movie2.columns, index = data_movie2.columns)
data_cosine
```

## Modelling Hybrid Recommender

## 1. Case Person "luck" Want to Watch Movie that Similar with "Captain Marvel"

### 1.1 Show Movies with High Similarity with "Captain Marvel"

```python
data_nama = pd.DataFrame(data_cosine.loc["Captain Marvel"].sort_values(ascending = False))
data_nama = data_nama.iloc[1:6]
data_nama
```

### 1.2 Show People Rating for Movies with High Similarity with "Captain Marvel"
* Step 1: Change the data column "Nama Anda" into index
* Step 2: Transpose the data and rename the axis to "Judul Film"
* Step 3: Join high similarity movies data with user rating data using the inner method
* Step 4: Transpose the data and rename the axis to "Nama Anda"

```python
data_rating.set_index("Nama Anda", inplace = True)
data_rating = data_rating.transpose()
data_rating = data_rating.rename_axis(index = "Judul Film")
data_re = pd.merge(data_nama, data_rating, how = "inner", on = "Judul Film")
data_re.drop("Captain Marvel", axis = 1, inplace = True)
data_re = data_re.transpose()
data_re = data_re.rename_axis(index = "Nama Anda")
data_re.head()
```

### 1.3 Calculate Each Person's Correlation

```python
data_ori = data_re
data_re = data_re.transpose()
data_corr = data_re.corr(method = "pearson")
data_corr
```

### 1.4 Make Matrix Rating from People with Correlation > 0 with "luck"
* Step 1: Take correlation data with the column "luck" and sorted from largest to smallest
* Step 2: Filter data by only taking correlation values that are more than 0
* Step 3: Join correlation data with user rating data using the inner method

```python
data_rec = pd.DataFrame(data_corr.loc["luck"].sort_values(ascending = False))
data_rec = data_rec[data_rec["luck"] > 0]
data_rec = pd.merge(data_rec, data_ori, how = "inner", on = "Nama Anda")
data_rec
```

### 1.5 Calculate Total Correlation Value
* Step 1: Transpose the data
* Step 2: Take data with the column "luck"
* Step 3: Do .sum() method to that column

```python
data_rec = data_rec.transpose()
data_corr_film = data_rec.iloc[0,1:]
data_corr_film = pd.DataFrame(data_corr_film)
sum_cor = data_corr_film["luck"].sum()
sum_cor
```

### 1.6 Show the Movie that Hasn't Been Watched by "luck"
* Step 1: Dropping the unused column
* Step 2: Take the movies title that was rated "0" by "luck"
* Step 3: Dropping the unused row

```python
data_rec.drop("luck", axis = 0, inplace = True)
data_rec = data_rec[data_rec["luck"] == 0]
data_rec.drop("luck", axis = 1, inplace = True)
data_rec
```

### 1.7 Show Recommended Movie with Predicted Rating
* Step 1: Do matrix multiplication to similarity data and people rating for movies that "luck" hasn't watched data
* Step 2: Divide the matrix multiplication results by total similarity value
* Step 3: Show the movies recommendation with the predicted rating value

```python
data_reccomend = data_rec.dot(data_corr_film)
data_reccomend["luck"] = data_reccomend["luck"] / sum_cor
data_reccomend.sort_values("luck", ascending = False)
```

***
Based on the results above, there are 4 recommended movies for "luck" if he wants to watch a movie that similar to "Captain Marvel" including Spiderman: Far From Home (3.270666), Aladdin (2.171136), Gundala (1.822844) and The Lion King (1.685959).

## 2. Case Person "bunga" Want to Watch Movie that Similar with "Dilan 1991"

### 2.1 Show Movies with High Similarity with "Dilan 1991"

```python
data_nama1 = pd.DataFrame(data_cosine.loc["Dilan 1991"].sort_values(ascending = False))
data_nama1 = data_nama1.iloc[1:6]
data_nama1
```

### 2.2 Show People Rating for Movies with High Similarity with "Dilan 1991"
* Step 1: Join high similarity movies data with user rating data using the inner method
* Step 2: Transpose the data and rename the axis to "Nama Anda"

```python
data_re1 = pd.merge(data_nama1, data_rating, how = "inner", on = "Judul Film")
data_re1.drop("Dilan 1991", axis = 1, inplace = True)
data_re1 = data_re1.transpose()
data_re1 = data_re1.rename_axis(index = "Nama Anda")
data_re1.head()
```

### 2.3 Calculate Each Person's Correlation

```python
data_ori = data_re1
data_re1 = data_re1.transpose()
data_corr1 = data_re1.corr(method = "pearson")
data_corr1
```

### 2.4 Make Matrix Rating from People with Correlation > 0 with "bunga"
* Step 1: Take correlation data with the column "bunga" and sorted from largest to smallest
* Step 2: Filter data by only taking correlation values that are more than 0
* Step 3: Join correlation data with user rating data using the inner method

```python
data_rec1 = pd.DataFrame(data_corr1.loc["bunga"].sort_values(ascending = False))
data_rec1 = data_rec1[data_rec1["bunga"] > 0]
data_rec1 = pd.merge(data_rec1, data_ori, how = "inner", on = "Nama Anda")
data_rec1
```

### 2.5 Calculate Total Correlation Value
* Step 1: Transpose the data
* Step 2: Take data with the column "bunga"
* Step 3: Do .sum() method to that column

```python
data_rec1 = data_rec1.transpose()
data_corr_film1 = data_rec1.iloc[0,1:]
data_corr_film1 = pd.DataFrame(data_corr_film1)
sum_cor1 = data_corr_film1["bunga"].sum()
sum_cor1
```

### 2.6 Show the Movie that Hasn't Been Watched by "bunga"
* Step 1: Dropping the unused column
* Step 2: Take the movies title that was rated "0" by "bunga"
* Step 3: Dropping the unused row

```python
data_rec1.drop("bunga", axis = 0, inplace = True)
data_rec1 = data_rec1[data_rec1["bunga"] == 0]
data_rec1.drop("bunga", axis = 1, inplace = True)
data_rec1
```

### 2.7 Show Recommended Movie with Predicted Rating
* Step 1: Do matrix multiplication to similarity data and people rating for movies that "bunga" hasn't watched data
* Step 2: Divide the matrix multiplication results by total similarity value
* Step 3: Show the movies recommendation with the predicted rating value

```python
data_reccomend1 = data_rec1.dot(data_corr_film1)
data_reccomend1["bunga"] = data_reccomend1["bunga"] / sum_cor1
data_reccomend1.sort_values("bunga", ascending = False)
```

***
Based on the results above, there are 3 recommended movies for "bunga" if she wants to watch a movie that similar to "Dilan 1991" including Ada Apa dengan Cinta 2 (2.511585), Dua Garis Biru (1.165119) dan Bumi Manusia (0.629451).
***

## Conclusion
By using Hybrid Recommender we can use two data as inputs, the inputs are the name of the person and the title of the movie that they are interested in. From that, we can give the movies recommendation that can be watched based on the similarity between the movie with the other movies and based on the correlation between the person with the other people. In the real case, we can apply this algorithm to display the movies recommendation on a movie website when someone has logged in and has searched for a movie on the website.
