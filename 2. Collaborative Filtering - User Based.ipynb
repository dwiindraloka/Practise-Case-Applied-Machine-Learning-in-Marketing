## Import library

```python
import pandas as pd
```

## Upload Dataset

```python
data = pd.read_csv("Film.csv")
data.head()
```

## Drop Unused Column

```python
data.drop("Timestamp", axis = 1, inplace = True)
```

## Change Column "Nama Anda" into Index

```python
col = data["Nama Anda"]
```

```python
data.set_index("Nama Anda", inplace = True)
data.head()
```

## Fill NaN Value with 0

```python
data.isna().sum()
```

```python
data = data.fillna(0)
data_ori = data
data.head()
```

## Calculate Each Person's Similarity Using Eucledian Similarity

### 1. Matrix Similarity Eucledian

```python
from scipy.spatial import distance_matrix
data = data.transpose()
data_euclid = pd.DataFrame(1 / (1 + distance_matrix(data.T, data.T)), columns = col, index = col)
data_euclid
```

### 2. Show People with High Similarity with "Topik Zulkarnain"

```python
data_nama = pd.DataFrame(data_euclid.loc["Topik Zulkarnain"].sort_values(ascending = False))
data_nama.iloc[1:6]
```

### 3. Show People with High Similarity with "Hania"

```python
data_nama = pd.DataFrame(data_euclid.loc["Hania"].sort_values(ascending = False))
data_nama.iloc[1:6]
```

## Modelling CF User Based Using Cosine Similarity

## 1. Case for Person "Topik Zulkarnain"

### 1.1 Make Matrix Rating from People with Similarity > 0.5 with "Topik Zulkarnain"
* Step 1: Take cosine similarity data with the column "Topik Zulkarnain" and sorted from largest to smallest
* Step 2: Filter data by only taking similarity values that are more than 0.5
* Step 3: Join similarity data with user rating data using the inner method

```python
data_re = pd.DataFrame(data_cosine.loc["Topik Zulkarnain"].sort_values(ascending = False))
data_re = data_re[data_re["Topik Zulkarnain"] > 0.5]
data_re = pd.merge(data_re, data_ori, how = "inner", on = "Nama Anda")
data_re
```

### 1.2 Calculate Total Similarity Value
* Step 1: Transpose the data
* Step 2: Take data with the column "Topik Zulkarnain"
* Step 3: Do .sum() method to that column

```python
data_re = data_re.transpose()
data_cosine_film = data_re.iloc[0,1:]
data_cosine_film = pd.DataFrame(data_cosine_film)
sum_cosine = data_cosine_film["Topik Zulkarnain"].sum()
sum_cosine
```

### 1.3 Show the Movie that Hasn't Been Watched by "Topik Zulkarnain"
* Step 1: Dropping the unused column
* Step 2: Take the movies title that was rated "0" by "Topik Zulkarnain"
* Step 3: Dropping the unused row

```python
data_re.drop("Topik Zulkarnain", axis = 0, inplace = True)
data_re = data_re[data_re["Topik Zulkarnain"] == 0]
data_re.drop("Topik Zulkarnain", axis = 1, inplace = True)
data_re
```

### 1.4 Show Recommended Movie with Predicted Rating
* Step 1: Do matrix multiplication to similarity data and people rating for movies that "Topik Zulkarnain" hasn't watched data
* Step 2: Divide the matrix multiplication results by total similarity value
* Step 3: Show the movies recommendation with the predicted rating value

```python
data_reccom = data_re.dot(data_cosine_film)
data_reccom["Topik Zulkarnain"] = data_reccom["Topik Zulkarnain"] / sum_cosine
data_reccom.sort_values("Topik Zulkarnain", ascending = False)
```
