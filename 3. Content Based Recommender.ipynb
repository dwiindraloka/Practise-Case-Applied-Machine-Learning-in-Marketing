## Import Library

```python
import pandas as pd
```

## Upload Dataset

```python
data = pd.read_csv("Film1.csv")
data.head()
```

## Change Column "Judul Film" into Index

```python
data.set_index("Judul Film", inplace = True)
data_ori = data
data.head()
```

## Make Data Dummies

```python
data = pd.get_dummies(data)
data.head()
```

## Normalize the Data Using MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler
data1 = data.transpose()
col = data.columns
scaler = MinMaxScaler().fit_transform(data)
data2 = pd.DataFrame(scaler, columns = col)
data2.head()
```

## Calculate Each Movie's Similarity Using Eucledian Similarity

```python
from scipy.spatial import distance_matrix
data2 = data2.transpose()
data_euclid = pd.DataFrame(1 / (1 + distance_matrix(data2.T, data2.T)), columns = data1.columns, index = data1.columns)
data_euclid
```

## Calculate Each Movie's Similarity Using Cosine Similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
data_cosine = cosine_similarity(data2.transpose())
data_cosine = pd.DataFrame(data_cosine, columns = data1.columns, index = data1.columns)
data_cosine
```

***
Based on the results above, we choose to use Cosine Similarity, because the results that obtained from Cosine Similarity provide a better distribution compared to Eucledian Similarity.
***

## Modelling Content Based Recommender

## 1. Content Based Recommender Using "Rating" and "Genre"

### 1.1 Make New Data With Coumns "Rating" and "Genre"

```python
column = ["Rating", "Genre"]
data_rg = data_ori[column]
data_rg.head()
```

### 1.2 Make Data Dummies

```python
data_rg = pd.get_dummies(data_rg)
data_rg.head()
```

### 1.3 Normalize the Data Using MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler
col1 = data_rg.columns
scaler = MinMaxScaler().fit_transform(data_rg)
data_rg = pd.DataFrame(scaler, columns = col1)
data_rg.head()
```

### 1.4 Calculate Each Movie's Similarity Using Cosine Similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
data_cosine1 = cosine_similarity(data_rg)
data_cosine1 = pd.DataFrame(data_cosine1, columns = data1.columns, index = data1.columns)
data_cosine1
```

### 1.5 Get Recommendation If We Like "Dilan 1991"

```python
data_nama = pd.DataFrame(data_cosine1.loc["Dilan 1991"].sort_values(ascending = False))
data_nama.iloc[1:4]
```

### 1.6 Get Recommendation If We Like "Gundala"

```python
data_nama = pd.DataFrame(data_cosine1.loc["Gundala"].sort_values(ascending = False))
data_nama.iloc[1:4]
```

***
Based on the results above, the recommended movies if we like "Dilan 1991" are Bumi Manusia, Ada Apa dengan Cinta 2 and Dua Garis Biru and the recommended movies if we like "Gundala" are Spiderman: Far From Home, Captain Marvel and Avengers: End Game.
***

## 2. Content Based Recommender Using "Negara", "Rating" and "Durasi"
